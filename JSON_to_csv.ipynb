```python
import os
import json
import pandas as pd

key_list = ["Bolnr", "IdocCreationDate", "Refnr", "Country1", "Belnr", "Inco2"]

def extract_values_iterative(json_obj, values_dic):
    stack = [json_obj]
    while stack:
        current_obj = stack.pop()
        if isinstance(current_obj, dict):
            for key, value in current_obj.items():
                if key in key_list:
                    if key in values_dic:
                        # Se a chave já existe no dicionário, adicione o valor à lista de valores
                        values_dic[key].append(value)
                    else:
                        # Caso contrário, crie uma nova lista com o valor
                        values_dic[key] = [value]
                stack.append(value)
        elif isinstance(current_obj, list):
            stack.extend(current_obj)

def get_keys_dic(values_dic, keys):
    keys_dic = {}
    for k in values_dic:
        if k in keys:
            # Se houver mais de um valor para a mesma chave, concatene-os em uma única string
            keys_dic[k] = ', '.join(str(val) for val in values_dic[k])
    return keys_dic

json_path = r'C:\Users\User\Desktop\EXPEDITORS\REI\REI - JSONs'
keys_dic_list = []

if os.path.exists(json_path) and os.path.isdir(json_path):
    for file in os.listdir(json_path):
        if file.endswith('.json'):
            file_path = os.path.join(json_path, file)
            with open(file_path, 'r', encoding='utf-8') as f:
                dados_json = json.load(f)
                values_dic = {}
                extract_values_iterative(dados_json, values_dic)
                keys_dic = get_keys_dic(values_dic, key_list)
                keys_dic['FileName'] = file
                keys_dic_list.append(keys_dic)
else:
    print("The path doesn't exist")

df = pd.DataFrame(keys_dic_list)

# Se quiser que cada valor da lista seja uma linha separada, use o método explode
df = df.apply(pd.Series.explode)

df
```
